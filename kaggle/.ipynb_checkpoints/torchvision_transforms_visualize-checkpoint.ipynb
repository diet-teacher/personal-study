```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import PIL
import torch
import numpy as np

from torchvision.transforms.functional import to_pil_image

def transform_practice(img_path, transform):
    plt.figure(figsize=(10, 4))
    for i in range(10):
        pil_image = PIL.Image.open(img_path)
        tensor = torch.as_tensor(np.asarray(pil_image)).permute(2,0,1)
        applicated_img = transform(tensor)
    
        plt.subplot(2, 5, i+1)
        plt.imshow(applicated_img.permute(1,2,0).numpy())
        plt.axis("off")
    plt.show()
```

```python
# permute(index)를 하면 그 위치로 reshape가 됨.

pil_image = PIL.Image.open('/Users/iseung-won/Downloads/IMG_9722.JPG')
print('none :', torch.as_tensor(np.asarray(pil_image)).size())
print('permute(2,0,1) :', torch.as_tensor(np.asarray(pil_image)).permute(2,0,1).size())
print('permute(1,2,0) :', torch.as_tensor(np.asarray(pil_image)).permute(1,2,0).size())
print('permute(1,0,2) :', torch.as_tensor(np.asarray(pil_image)).permute(1,0,2).size())
```

```python
# transforms.RandAugment()

transform = transforms.RandAugment()
img_path = '/Users/iseung-won/Downloads/IMG_9722.JPG'

transform_practice(img_path=img_path, transform=transform)
```

```python
# transforms.RandomAdjustSharpness

img_path = '/Users/iseung-won/Downloads/IMG_9722.JPG'

transform = transforms.RandomAdjustSharpness(0,p=1)
transform1 = transforms.RandomAdjustSharpness(1,p=1)
transform2 = transforms.RandomAdjustSharpness(10,p=1)
transform3 = transforms.RandomAdjustSharpness(30,p=1)

pil_image = PIL.Image.open(img_path)
tensor = torch.as_tensor(np.asarray(pil_image)).permute(2,0,1)
applicated_img = transform(tensor)
applicated_img1 = transform1(tensor)
applicated_img2 = transform2(tensor)
applicated_img3 = transform3(tensor)

plt.figure(figsize=(10, 13))
plt.subplot(2, 2, 1)
plt.imshow(applicated_img.permute(1,2,0).numpy())
plt.axis("off")
plt.title('Sharpness(0)')
plt.subplot(2, 2, 2)
plt.imshow(applicated_img1.permute(1,2,0).numpy())
plt.axis("off")
plt.title('Sharpness(1)')
plt.subplot(2, 2, 3)
plt.imshow(applicated_img2.permute(1,2,0).numpy())
plt.axis("off")
plt.title('Sharpness(10)')
plt.subplot(2, 2, 4)
plt.imshow(applicated_img3.permute(1,2,0).numpy())
plt.axis("off")
plt.title('Sharpness(30)')
```

```python
# transforms.RandomHorizontalFlip

transform = transforms.RandomHorizontalFlip(p=0.5)
img_path = '/Users/iseung-won/Downloads/IMG_9722.JPG'

transform_practice(img_path=img_path, transform=transform)
```

```python
# transforms.RandomAffine(180)
# RandomRotation과 비슷하지만 shear argument로 원본에서 외곡을 만들수 있음

transform = transforms.RandomAffine(180,shear=45)
img_path = '/Users/iseung-won/Downloads/IMG_9722.JPG'

transform_practice(img_path=img_path, transform=transform)
```
