```python
%cd
%cd vision/references/detection/
import utils
import transforms as T
def get_transform(train):
    transforms = []
    transforms.append(T.PILToTensor())
    transforms.append(T.ConvertImageDtype(torch.float))
    if train:
        transforms.append(T.RandomHorizontalFlip(0.5))

    return T.Compose(transforms)
```

```python
# 간단한 data_set 만들기 class 정의

from torch.utils.data import Dataset, DataLoader, random_split
import torch

class MNIST_dataset(Dataset):
    def __init__(self, img, cls, transforms):
        self.img = img
        self.cls = cls
        self.transforms = transforms
        
    def __getitem__(self, idx):
        
        img = Image.fromarray(self.img[idx].astype('uint8'), mode='L')
        cls = self.cls[idx]
        
        if self.transforms is not None:
            img, cls = self.transforms(img, cls)

        return img, cls
    
    def __len__(self):
        return self.img.shape[0]
    
```

```python
train_dataset = MNIST_dataset(img=x_t, cls=y_t, transforms=get_transform(train=True))
len(train_dataset)
train_loader = DataLoader(dataset=train_dataset,
                         batch_size=4,
                         shuffle=True)
```
